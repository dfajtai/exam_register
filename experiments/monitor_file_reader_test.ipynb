```python
import re
import requests
import lxml.html
import pprint

import sys

import platform

import chardet

print(sys.version)

platform.python_version()
```

```python
main_version = ".".join(str(platform.python_version()).split(".")[:-1])
main_version
```

```python
encodings = None

url = f"https://docs.python.org/{main_version}/library/codecs.html#standard-encodings"
html = requests.get(url).text
# Work-around for weird HTML markup in recent versions of Python documentation:
html = re.sub('<[/]?p>', '', html)
doc = lxml.html.fromstring(html)
standard_encodings_table = doc.xpath(
    '//table[preceding::h2[.//text()[contains(., "Standard Encodings")]]][//th/text()="Codec"]'
)[0]
encodings = standard_encodings_table.xpath('.//td[1]/text()')
```

```python
file = "monitor_center_file_sample/Trend20231005_8_1.bs"

raw_data_dict = {}

for e in encodings:
    _err = 0
    try:
        rawdata = open(file, "r", encoding=e).read()
        _err+=1
        raw_data_dict[e] = rawdata
        
        result = chardet.detect(rawdata)
        _err+=1
        charenc = result['encoding']
        
        print(f"{e} -> {charenc}")
    except Exception as exc:
        if _err == 0:
            print(f"{e} -> 'read()' failed")
        elif _err == 1:
            print(f"{e} ->  'chardet.detect()' failed")
        else:
            print(exc)
        continue
```

```python
for encoding, data in raw_data_dict.items():
    print(encoding)
    print(data[:1000])
```

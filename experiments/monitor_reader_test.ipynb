```python
def retro_pdf_data_extract_single_page(pdf_path,
                                       page = 1,
                                       table_count = 1,
                                       metadata_rows = 2, data_start_row = 5, data_end_row = -1,
                                       colnames = None):
    
    assert os.path.exists(pdf_path)
    tabula_tables = tabula.read_pdf(pdf_path, pages = page)
    
    if len(tabula_tables) < table_count:
        return
    
    res = []
    
    for taula_table in tabula_tables:
        # parse columns
        
        try:
            if isinstance(colnames,type(None)):
                tabula_raw_cols = taula_table[metadata_rows:data_start_row]
                #print(tabula_raw_cols)
                parsed_cols = []

                for c in list(tabula_raw_cols.columns):   
                    data = tabula_raw_cols[c].tolist()
                    data = [str(d) for d in data if not pd.isna(d)]
                    data = "".join(data)
                    for d in data.split(" "):
                        if d == "":
                            continue
                        parsed_cols.append(d)

                print("Columns parsed")

                #print(parsed_cols)
            else:
                parsed_cols = list(colnames)
            #print(parsed_cols)
            
            # parse dataframe
            
            df = taula_table[data_start_row:data_end_row].reset_index(drop=True)
            
            if len(df.columns.tolist()) != parsed_cols:
                # split columns on space                
                _df = pd.DataFrame()
                new_col_index = 0
                time_col_found = False
                for c in df.columns.tolist():
                    try:
                        if not time_col_found:
                            _col = pd.to_datetime(df[c])
                            _df[c] = _col
                            time_col_found = True
                            continue
                        
                    except ValueError:
                        pass
                    
                    if df[c].apply(lambda x: " " in str(x).strip()).all():
                        _col = df[c].apply(lambda x: str(re.sub(' +', ' ',str(x))))
                        c1 = f"New_col_{new_col_index}"
                        c2 = f"New_col_{new_col_index+1}"
                        new_col_index+=2
                        _df[[c1,c2]] = _col.str.split(' ', expand=True)
                    else:
                        _df[c] = df[c]                        
                df = _df
            
            assert(len(df.columns.tolist()) == len(parsed_cols))
            df.columns = parsed_cols
            
            print("Dataframe parsed")
            
            # parse metadata
            metadata_dict = {}
            if metadata_rows > 0:
                metadata_rows = taula_table[:metadata_rows]
                
                metadata_candidates = []
                metadata_candidates.extend(metadata_rows.columns.tolist())
                for index, row in metadata_rows.iterrows():
                    metadata_candidates.extend(row.tolist())

                #print(metadata_rows)

                for mc in metadata_candidates:
                    if pd.isna(mc):
                        continue
                    if "Unnamed" in mc:
                        continue
                    mc_data = str(mc).replace(": ","#").split(" ")
                    for _ in mc_data:
                        _mc_data = _.split("#")
                        metadata_dict[_mc_data[0]] = " ".join(_mc_data[1:]).strip()

                print("Metadata parsed")
            
            res.append({"metadata":metadata_dict,"dataframe":df}) 
        
        except Exception as ex:
            print(ex)
            continue
        
        
    return res
    
```

```python
for p in range(num_of_pages):
    print(f"page {p+1}")
    res = retro_pdf_data_extract_single_page(pdf_path=sample_path)
    if res:
        for r in res:
            print(r.get("metadata"))
            print(r.get("dataframe"))
    
```

```python
import glob

samples = glob.glob("monitor_retro_pdf_sample/*.pdf")

default_cols = ['idő', 'HR(BPM)', 'PR(BPM)', 'SpO2(%)', 'RR(RPM)', 'NIBP(mmHg)']

for s in samples:
    print(s)
    
    pdf_data = PdfReader(s)
    
    n = len(pdf_data.pages)
    
    for p in range(n):
        print(f"page {p+1}")
        res = retro_pdf_data_extract_single_page(pdf_path=s, colnames=default_cols )
        if res:
            for r in res:
                print(r.get("metadata"))
                print(r.get("dataframe"))
```
